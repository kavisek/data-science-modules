# Create Dummy Variables

<span>Pandas to the rescue again. The pandas "pd.to_dummies" function can be operated on dataframe to take one categorical feature and create a one-hot-encoded feature for each dimension in the original feature. The great part is that pandas will complete this actions for all object and categorical feature at one. So you can go from a dataframe 11 feature to 45 pretty quickly.</span>

### Import Preliminaries

```python
# Import modules
import pandas as pd

# Read in pokemon data 
df = pd.read_csv('Data/pokemon.csv', index_col='#')

# Drop name feature
df = df.drop('Name', axis=1)

# Print the numbef of feature in dataframe
print('Number of Features in DataFrame:', df.shape[1])
```

```python
# View the head of the dataframe
df.head()
```

```python
# Get dummries for non ordinal valuse in the dataframe
df = pd.get_dummies(df)

# Print the new number of features in the dataframe
print('Number of Features in DataFrame:', df.shape[1])
```

```python
# View the head of the dataframe
df.head(5)
```
