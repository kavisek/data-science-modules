### Scaling

Scaling is just resizing of the image. OpenCV comes with a function cv2.resize() for this purpose. The size of the image can be specified manually, or you can specify the scaling factor. Different interpolation methods are used. Preferable interpolation methods are cv2.INTER_AREA for shrinking and cv2.INTER_CUBIC (slow) & cv2.INTER_LINEAR for zooming. By default, interpolation method used is cv2.INTER_LINEAR for all resizing purposes. You can resize an input image either of following methods:

```python
import cv2
import numpy as np
import matplotlib.pylab as plt
img = cv2.imread('Data/messi1.jpg')
plt.imshow(res) ; plt.show()
```

```python
res = cv2.resize(img, None, fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
plt.imshow(res) ; plt.show()
```

```python
# Another Way to do this
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
```

### Translation

Translation is the shifting of objects location. If you know the shift in (x,y) direction, let it be (t_x,t_y), you can create the transformation matrix  \textbf{M} as follows:



```python
import cv2
import numpy as np

img = cv2.imread('Data/messi1.jpg', 0)
rows, cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (cols, rows))

plt.imshow(dst); plt.show()
```

### Rotation

```python
img = cv2.imread('Data/messi1.jpg', 0)
plt.imshow(img); plt.show()
```

```python

rows, cols = img.shape

M = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 1)
dst = cv2.warpAffine(img, M, (cols, rows))
plt.imshow(dst); plt.show()
```

### Affline Transformation

```python
img = cv2.imread('Data/drawing.png')
plt.imshow(img); plt.show()
```

```python
rows, cols, ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10, 100],[200,50],[100, 250]])

M = cv2.getAffineTransform(pts1, pts2)

dts = cv2.warpAffine(img, M,(cols, rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dts),plt.title('Output')
plt.show()
```

### Perspective Transformation

```python
img = cv2.imread('Data/sudokusmall.png')
plt.imshow(img); plt.show()
```

```python
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
