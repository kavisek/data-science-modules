# Index, Select, And Filter pandas Dataframes
### Import modules

```python
import pandas as pd
```

### Create a dataframe

```python
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [4, 24, 31, 2, 3],
        'coverage': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
df
```

### View a column of the dataframe

```python
df['name']
```

### View two columns of the dataframe

```python
df[['name','reports']]
```

### View the first two rows of the dataframe

```python
df[:2]
```

### View all rows where coverage is more than 50

```python
df[df['coverage']>50]
```

### View a row

```python
df.loc['Maricopa']
```

```python
df.loc[:,'coverage']
```

```python
df.loc['Yuma','coverage']
```
