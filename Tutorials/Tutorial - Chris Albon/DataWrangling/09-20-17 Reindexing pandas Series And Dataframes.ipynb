# Reindexing pandas Series and DataFrames
### Series

```python
# import modules
import pandas as pd
import numpy as np
```

```python
# Create a pandas series of the risk of fire in Southern Arizona
```

```python
brushFireRisk = pd.Series([34,23,12,23], 
                          index=['Bisbee', 'Douglas', 
                                 'Sierra Vista', 'Tombstone'])
brushFireRisk
```

```python
# Reindex the series and create a new series variable
brushFireRiskReindexed = brushFireRisk.reindex(['Tombstone', 'Douglas', 'Bisbee', 
                                                'Sierra Vista', 'Barley', 'Tucson'])
brushFireRiskReindexed
```

```python
# Reindex the series and fill in any missing indexes as 0
brushFireRiskReindexed = brushFireRisk.reindex(['Tombstone','Douglas',
                                               'Bisbee', 'Sierra Vista', 
                                               'Barley', 'Tucson'],
                                               fill_value=0)
brushFireRiskReindexed
```

# DataFrames

```python
# C# Create a dataframe
data = {'county': ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [4, 24, 31, 2, 3]}
df = pd.DataFrame(data)
df
```

```python
# Change the order (the index) of the rows
df.reindex([4,3,2,1,0])
```

```python
# Change the orde (the index) of the columns
columnsTitles = ['year', 'reports', 'county']
df.reindex(columns=columnsTitles)
```
