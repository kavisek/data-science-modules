# Select Rows When Columns Contain Certain Values
### Preliminaries

```python
# Import modules
import pandas as pd

# Set ipython's max row display
pd.set_option('display.max_rows', 1000)

# set ipython's max column width to 50
pd.set_option('display.max_columns', 50)
```

### Create an example dataframe

```python
# Create an example dataframe
data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'year': [2012, 2012, 2013, 2014, 2014], 
        'reports': [4, 24, 31, 2, 3]}
df = pd.DataFrame(data, index = ['Cochice', 'Pima', 'Santa Cruz', 'Maricopa', 'Yuma'])
df
```

### Grab rows based on column values

```python
value_list =  ['Tina', 'Molly', 'Jason']
```

```python
# Grab dataFrame row where column has certain values
df[df.name.isin(value_list)]
```

```python
# Graba DataFrame rows where column doesn't have certain values
df[~df.name.isin(value_list)]
```
