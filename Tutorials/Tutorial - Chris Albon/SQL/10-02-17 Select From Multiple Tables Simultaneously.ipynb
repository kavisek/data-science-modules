### View All Unique City Names From Both Tables

```python
%%sql

SELECT city FROM criminals
UNION
SELECT city FROM crimes;
```

### View All City Names From Both Tables

```python
%%sql
SELECT city FROM criminals
UNION ALL
SELECT city FROM crimes;
```
