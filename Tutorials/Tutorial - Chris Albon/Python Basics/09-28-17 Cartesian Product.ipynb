# Cartesian Product
### Preliminaries

```python
# import pandas as pd
import pandas as pd
```

### Create Data

```python
# Create two lists
i = [1,2,3,4,5]
j = [1,2,3,4,5]
```

###  Calculate Cartesian Product (Method 1)

```python
# List every single x in i with every single y (i.e. Cartesian product)
[(x,y) for x in i for y in j ]
```

### Calculate Cartesian Product (Method 2)

```python
# An alternative way to do the cartesian product

# import itertools
import itertools

# for two sets, find the the cartisan produc
for i in itertools.product([1,2,3,4],[1,2,3,4,5]):
    # and print it
    print(i)
```
