```python
# Import modules
import pandas as pd
from sklearn.decomposition import FactorAnalysis
from sklearn.preprocessing import LabelEncoder

# Encoding Functions
def EncodeDataFrame(df):
  '''encode dataframe values'''
  object_columns = list(df.select_dtypes( include=[object]).columns)
  for col in df[object_columns]:
    df[col] = LabelEncoder().fit_transform(df[col])
  return df

# Styling Function
def color_palette(val):
  color = 'red' if val < 0 else 'green'
  color = 'black' if val == 0
  return 'color: %s'%color
```

```python
# Import data
train_df = pd.DataFrame(pd.read_csv('https://www.dropbox.com/s/m2xk1588efn3icv/cleaned_train_df.csv?dl=1'))
test_df = pd.DataFrame(pd.read_csv('https://www.dropbox.com/s/zlq1xsznwg2a7z6/cleaned_test_df.csv?dl=1'))

# Label Encoding Feature Data
EncodeDataFrame(train_df)

# Convert dataframe to values
X = train_df[1:].values
y = train_df['Survived'].values 

# Factor analysis
factor = FactorAnalysis(n_components=train_df.shape[1], random_state=10).fit(X)
factor = pd.DataFrame(factor.components_ , columns=train_df.columns)

factor.style.applymap(color_palette)
```
